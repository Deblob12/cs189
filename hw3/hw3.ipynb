# 2 Isocontours of Normal Distributions

```python
fig_num2 = 0
def isocontour_plotter1(mean, covariance, title):
    global fig_num2
    random_variable = multivariate_normal(mean, covariance)
    x, y = np.mgrid[-1:1:.01, -1:1:.01]
    pos = np.dstack((x, y))
    fig_num2 += 1
    plt.figure(fig_num2)
    plt.title(title)
    plt.contourf(x, y, random_variable.pdf(pos))   

def isocontour_plotter2(means, covariances, title):
    global fig_num2
    random_variable1 = multivariate_normal(means[0], covariances[0])
    random_variable2 = multivariate_normal(means[1], covariances[1])
    x, y = np.mgrid[-1:1:.01, -1:1:.01]
    pos = np.dstack((x, y))
    fig_num2 += 1
    plt.figure(fig_num2)
    plt.title(title)
    plt.contourf(x, y, 
                 random_variable1.pdf(pos)-random_variable2.pdf(pos), 20)  
    
isocontour_plotter1([1,1], [[1,0],[0,2]], "2a")
isocontour_plotter1([-1,2],[[2,1],[1,3]], "2b")
isocontour_plotter2([[0,2],[2,0]],[[[2,1],[1,1]],[[2,1],[1,1]]], "2c")
isocontour_plotter2([[0,2],[2,0]],[[[2,1],[1,1]],[[2,1],[1,3]]], "2d")
isocontour_plotter2([[1,1],[-1,-1]],[[[2,0],[0,1]],[[2,1],[1,2]]], "2e")
```

# 3 Eigenvectors of the Gaussian Covariance Matrix

```python
X1 = np.random.normal(3.0,3.0,100) #sd = sqrt(var)
X2_rv = np.random.normal(4.0,2.0,100)
samples = [[x1,(0.5*x1)+x2] for x1,x2 in zip(X1,X2_rv)]
# X2 = [sample[1] for sample in samples]
samples = np.array(samples)
```

```python
mean_3a = np.mean(samples,axis=0)
print("3a mean:",mean_3a)
```

```python
cov_3b = np.cov(samples,rowvar=False,bias=True)
print("3b covariance matrix:",cov_3b)
```

```python
eigvals, eigvecs = np.linalg.eig(cov_3b)
print("3c:")
print("eigenvectors:",eigvecs)
print("eigenvalues:",eigvals)
```

```python
plt.axis((-15,15,-15,15))
plt.title("3d")
plt.scatter(samples[:,0],samples[:,1])
plt.plot(eigvecs,color="green");
```

```python
rotated_points = (cov_3b.T @ (samples - mean_3a).T).T
plt.axis((-15,15,-15,15))
plt.title("3e")
plt.scatter(rotated_points[:,0],rotated_points[:,1]);
```

```python
#from hw
def shuffle_train_val_split(name, val_amt=0, percent=0):
    data = loadmat(name+"-data/"+name+"_data.mat")
    total_num_ex = data["training_data"].shape[0]
    shuffle = np.random.permutation(total_num_ex) #shuffle before split
    data_xtrain, data_ytrain = data["training_data"][shuffle], \
    data["training_labels"][shuffle] #split
    if not val_amt: #split by fixed number or percentage
        val_amt = int(percent * total_num_ex)
    data_xval, data_yval = data_xtrain[:val_amt,:], data_ytrain[:val_amt,:]
    data_xtrain, data_ytrain = data_xtrain[val_amt:,:], data_ytrain[val_amt:,:]
    return data_xtrain, data_ytrain, data_xval, data_yval
```

```python
def results_to_csv(y_test, setname):
    y_test = y_test.astype(int)
    df = pd.DataFrame({'Category': y_test})
    df.index += 1  # Ensures that the index starts at 1. 
    df.to_csv(setname+'-'+str(datetime.datetime.now().time())+ \
              '-'+'submission.csv', index_label='Id')
```

```python
#train/test split
mnist_xtrain, mnist_ytrain, mnist_xval, mnist_yval = \
shuffle_train_val_split("mnist", val_amt=10000) 

#preprocess
mnist_xtrain = (mnist_xtrain-np.mean(mnist_xtrain))/np.std(mnist_xtrain)
mnist_xval = (mnist_xval-np.mean(mnist_xval))/np.std(mnist_xval)

spam = loadmat("spam-data/spam_data.mat")
spam_xtrain, spam_ytrain, spam_xtest = \
spam["training_data"], spam["training_labels"], spam["test_data"]
```

```python
def probs_calc_LDA(train_amt=mnist_xtrain.shape[0],dataset="mnist"):
    means_7c = []
    covs_7c = []
    masks_7c = []
    for i in range(10): #0-9 digits
        mask, est_mean, est_covar = None, None, None
        if dataset == "mnist":
            mask = np.where(mnist_ytrain[:train_amt,:] == i)[0]
            est_mean = np.mean(mnist_xtrain[mask], axis=0)
            est_covar = np.cov(mnist_xtrain[mask], rowvar=False, bias=True)
        if dataset == "spam":
            mask = np.where(spam_ytrain[:train_amt,:] == i)[0]
            est_mean = np.mean(spam_xtrain[mask], axis=0)
            est_covar = np.cov(spam_xtrain[mask], rowvar=False, bias=True)
        means_7c.append(est_mean)
        covs_7c.append(est_covar)
        masks_7c.append(mask)
    covar_LDA = sum(covs_7c)/10.0 #one covariance matrix for all classes
    priors_7c = np.array([np.sum(mask)/n for mask in masks_7c])
    return means_7c, covs_7c, priors_7c, covar_LDA

def fit_gaussians_LDA(means, covar):
    LDA_gaussians = []
    for mean in means:
        LDA_gaussians.append(multivariate_normal(mean, covar, allow_singular=True))
    return LDA_gaussians

def generate_predictions_LDA(samples, gaussians, priors):
    pdfs = np.array([gauss.logpdf(samples) for gauss in gaussians]).T
    posteriors = pdfs + np.log(priors)
    predictions = posteriors.argmax(axis=1)
    return predictions
```

```python
def train_LDA(amts=[mnist_xtrain.shape[0]],dataset="mnist",mode="train"):
    train_errs, val_errs, gaussians, priors = [],[],[],[]
    for amt in amts:
        if dataset == "mnist":
            means, covs, priors, covar_LDA = probs_calc_LDA(amt)
            gaussians = fit_gaussians_LDA(means, covar_LDA)
            train_predictions = \
            generate_predictions_LDA(mnist_xtrain, gaussians, priors).reshape(*mnist_ytrain.shape)
            if mode == "train":
                val_predictions = \
                generate_predictions_LDA(mnist_xval, gaussians, priors).reshape(*mnist_yval.shape)
                train_errs.append(np.sum(train_predictions != mnist_ytrain)/mnist_ytrain.shape[0])
                val_errs.append((np.sum(val_predictions != mnist_yval)/mnist_yval.shape[0]))
        elif dataset == "spam":        
            means, covs, priors, covar_LDA = probs_calc_LDA(amt,"spam")
            gaussians = fit_gaussians_LDA(means, covar_LDA)
            train_predictions = \
            generate_predictions_LDA(spam_xtrain, gaussians, priors).reshape(*spam_ytrain.shape)
            if mode == "train":
                val_predictions = \
                generate_predictions_LDA(spam_xval, gaussians, priors).reshape(*spam_yval.shape)
                train_errs.append(np.sum(train_predictions != spam_ytrain)/spam_ytrain.shape[0])
                
    if dataset == "mnist" and mode == "train":
        plt.plot(amts,train_errs)
        plt.plot(amts,val_errs)
        plt.legend(["training error","validation error"])
        plt.title("LDA on MNIST")
        plt.xlabel("training size")
        plt.ylabel("error")
    return gaussians, priors
```

```python
amts = [100, 200, 500, 1000, 2000, 5000, 10000, 30000, 50000]
gaussians, priors = train_LDA(amts)
# for each training point:
#  10 pdf scores
#  10 priors
```
